```python
import numpy as np
import matplotlib.image as Image
import matplotlib.pyplot as plt
```

```python
path = './pictures/cloud.jpg'
```

```python
img = Image.imread(path)
```

```python
norm_err = np.sum(np.square(img))
```

```python
img.setflags(write=1)
```

```python
plt.imshow(img)
```

```python
img[img>127] = 255 - img[img>127]
```

```python
plt.imshow(img)
```

```python
max_err = np.sum(np.square(img))
```
